```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Create empty lists to store data
titles = []
years = []
engines = []
kms = []
dates = []
links = []
markas = []  # Added list for Marka values
phone_numbers = []

# Extracting information from each short-info block
for short_info in my_soup.find_all('div', class_='short-area'):
    # Extracting details
    # Inside the loop
    title = short_info.find('div', class_='pub-title').find('a').get_text(strip=True)
    titles.append(title)
    link = short_info.find('div', class_='short-img').find('a')['href']
    links.append(link)
    phone_number = None
    marka_value = None  # Added variable for Marka value

    if link:
        linked_response = requests.get(link)
        linked_soup = BeautifulSoup(linked_response.content, 'html.parser')

        spec_area = linked_soup.find('div', class_='spec-area')
        marka_value = spec_area.get_text(strip=True) if spec_area else None

    # Fetching the linked page
    if link:
        linked_response = requests.get(link)
        linked_soup = BeautifulSoup(linked_response.content, 'html.parser')

        # Extracting phone number
        phone_tag = linked_soup.find('a', {'class': 'phone-number'})  # Replace with the actual class or identifier
        phone_number = phone_tag.get_text(strip=True) if phone_tag else None

    markas.append(marka_value)  # Append Marka value to the list
    years.append(short_info.find('div', class_='xitems').find('span').get_text(strip=True))
    engines.append(short_info.find('div', class_='xitems').find_all('span')[1].get_text(strip=True))
    km_elem = short_info.find('div', class_='xitems').find_all('span')[2].find('span')
    kms.append(km_elem.get_text(strip=True) if km_elem else None)
    dates.append(short_info.find('div', class_='right-p').get_text(strip=True))
    phone_numbers.append(phone_number)

# Create DataFrame
df = pd.DataFrame({
    'Title': titles,
    'Link': links,
    'Marka': markas,
    'Year': years,
    'Engine': engines,
    'KM': kms,
    'Date': dates,
    'Phone': phone_numbers
})

# Print or use the DataFrame as needed
print(df)
```
